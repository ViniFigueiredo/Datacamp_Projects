```python
import pandas as pd
import matplotlib.pyplot as plt
netflix_df = pd.read_csv("netflix_data.csv")
```

# APLICANDO FILTROS AO DATAFRAME

```python
#Filmes lançados na década de 90
filmes_1990 = netflix_df[(netflix_df["type"] == "Movie") & 
                          (netflix_df["release_year"] >= 1990) & 
                          (netflix_df["release_year"] < 2000)]
```

# Construindo o Histograma

```python
plt.hist(filmes_1990["duration"])
plt.title('Histograma da duração dos filmes dos anos 90')
plt.xlabel('Duração (minutos)')
plt.ylabel('Quantidade de Filmes')
plt.show()
duration = 100
filmes_1990_acao = filmes_1990[filmes_1990["genre"] == "Action"]
```

# Quantidade de filmes de ação com menos de 1 hora e 30 minutos

```python
contagem_de_filmes_curtos = 0

for label, row in filmes_1990_acao.iterrows():
    if row["duration"] < 90:
        contagem_de_filmes_curtos = contagem_de_filmes_curtos + 1

print(contagem_de_filmes_curtos)
```
